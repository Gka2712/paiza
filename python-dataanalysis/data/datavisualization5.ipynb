```python
import pandas as pd
import seaborn as sns
df=pd.DataFrame({"a":pd.Series([1,2,3,5,4]),"b":pd.Series([3,5,4,2,1])})
sns.lineplot(data=df,x=df.index,y="a")
```

```python
sns.barplot(data=df,x="a",y="b")
```

```python
sns.barplot(data=df,x="a",y="b",orient="h")
```
