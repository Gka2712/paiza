```python
%matplotlib inline
import pandas as pd
b1=pd.Series([1,4,3,-2,6])
b1.plot()
```

```python
b2=pd.Series([3,2,1,5,-4])
df=pd.DataFrame({"b1":b1,"b2":b2})
df.plot()
```

```python
ind=["A","B","C","D","E"]
b3=pd.Series([3,2,5,4,1],index=ind)
b3.plot.bar()
```

```python
b4=pd.Series([1,4,2,3,5],index=ind)
df2=pd.DataFrame({"b3":b3,"b4":b4})
df2.plot.bar()
```

```python
b3.plot.barh()
```

```python
df2.plot.barh()
```
